```python
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
#SimpleImputer
impute_numeric = KNNImputer() #strategy= 'mean'no es necesario al tener parametros por deffault
numeric_columns = df.select_dtypes(include='number').columns
numeric_columns
```

```python
df[numeric_columns] = impute_numeric.fit_transform(df[numeric_columns])
```

```python
impute_categorico = SimpleImputer(strategy='most_frequent')

categorical_columns = df.select_dtypes(include= 'object').columns

df[categorical_columns] = impute_categorico.fit_transform(df[categorical_columns])
```
