```python
expirience = [1,2,3,4,5,6]
income_jr_k = [20,22,28,34,38,40]
income_chief_k = [40,44,48,56,59,60]
```

```python
plt.plot(expirience,income_jr_k)
plt.plot(expirience,income_chief_k)
plt.title('salario')
plt.xlabel('Experiencia')
plt.ylabel('salario')
```

```python
plt.plot(expirience,income_jr_k, color = 'red')
plt.plot(expirience,income_chief_k, color = 'black')
plt.title('salario')
plt.xlabel('Experiencia')
plt.ylabel('salario')
plt.legend(['Jr engineer', 'Chief engineer'])
```

```python
plt.plot(expirience,income_jr_k, color = 'red', label = 'Jr', linestyle = '--', marker = 'o')
plt.plot(expirience,income_chief_k, color = 'black', label = 'Chief', linestyle = ':', marker = '*')
plt.title('salario')
plt.xlabel('Experiencia')
plt.ylabel('salario')
plt.legend()
plt.grid(True)
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
categorias = ['a','b','c','d']
valores = [5,7,3,8]

datos_cajas = [np.random.normal(0,std,100) for std in range(1,4)]
datos_hist = np.random.rand(10_000)
```

```python
fig, axs = plt.subplots(2,2, figsize = (15,10))
#primer cuadrante
axs[0,0].plot(x,y1)
axs[0,0].plot(x,y2)
axs[0,0].set_title('grafico de lineas')

#segundo cuadrante
axs[0,1].bar(categorias,valores)
axs[0,1].set_title('grafico de barras')

#tercer cuadrante
axs[1,0].boxplot(datos_cajas)
axs[1,0].set_title('cajas y bigotes')

#cuarto cuadrante
axs[1,1].hist(datos_hist)
axs[1,1].set_title('histograma')
```
